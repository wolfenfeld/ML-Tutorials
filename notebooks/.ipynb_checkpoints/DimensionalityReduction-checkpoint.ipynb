# Dimentionalty Reduction

```python
import pandas as pd
import numpy as np
from numpy.linalg import eig, svd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
class DimensionalityReducer(object):
    
    def __init__(self, data, target_dimension):
       
        self.data=data
        self.target_dimension = target_dimension
        self.fit()
        
    def fit(self):
        raise NotImplementedError
    
    def transform(self, data_point):
        raise NotImplementedError
            
    def reconstruct_data(self, transformed_data_point, mean):
        raise NotImplementedError
        
```

```python
class PCADimensionalityReducer(DimensionalityReducer):
    
    def __init__(self,data,target_dimension):
        
        self.eigen_values = None 
        self.eigen_vectors = None
        
        super(PCADimensionalityReducer,self).__init__(data=data,target_dimension=target_dimension)
        
    def fit(self):
        self.eigen_values, self.eigen_vectors = eig(np.cov(self.data,rowvar=False))
        self.eigen_vectors = np.real(self.eigen_vectors)
        
        
    def transform(self, data_point):
        return np.dot(data_point - data_point.mean(), self.eigen_vectors[:,:self.target_dimension])
 
    def reconstruct_data(self, transformed_data_point, mean):
        return np.dot(transformed_data_point, np.transpose(self.eigen_vectors[:,:self.target_dimension])) + mean
```

```python
dataset = datasets.fetch_olivetti_faces(shuffle=True)
faces = dataset.data
```

```python
plt.imshow(faces[0].reshape((64,64))) 
plt.show()
```

```python
target_dim = 200

dr_pca = PCADimensionalityReducer(data=faces[:350],target_dimension=target_dim)
```

```python
mean = faces[351].mean()
transformed = dr_pca.transform(data_point=faces[351])
reconstructed_image_pca = dr_pca.reconstruct_data(transformed, mean).reshape((64,64))
```

```python
plt.imshow(reconstructed_image_pca) 
plt.show()
```

```python
class SVDDimensionalityReducer(DimensionalityReducer):
    
    def __init__(self,data, target_dimension):
        
        self.s = None 
        self.v = None
        self.u = None

        super(SVDDimensionalityReducer,self).__init__(data=data,target_dimension=target_dimension)

    
    def fit(self):
        self.u, self.s, self.v = svd(self.data)
        
    def transform(data_point):
            
    def reconstruct_data(self, target_dimension):

        s = np.diag(self.s[:target_dimension])
        u = self.u[:,:target_dimension]
        v = self.v[:target_dimension]

        return pd.DataFrame(np.dot(u, np.dot(s, v)))
        
```

```python
dr_svd = SVDDimensionalityReducer(data=pd.DataFrame(faces))

reconstructed_image_svd = dr_svd.reconstruct_data(target_dim).iloc[0].values.reshape((64,64))
```
